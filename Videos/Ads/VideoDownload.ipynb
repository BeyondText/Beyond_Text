```python
#excitement
import os
import json
from pytube import YouTube

# Change directory to the specified path
os.chdir('C:\\Users\\yjpan\\My Drive\\JAMS 3rd prep\\Video')

# Read the JSON file with exciting scores
with open("exciting_scores.json", "r") as json_file:
    exciting_scores = json.load(json_file)

# Create folders for 'exciting' and 'non-exciting' videos if they don't exist
if not os.path.exists("exciting"):
    os.makedirs("exciting")

if not os.path.exists("non_exciting"):
    os.makedirs("non_exciting")

# Initialize variables
counter = 0
max_videos = 300
video_ids = list(exciting_scores.keys())

# Download options for youtube_dl for getting video info
ydl_opts_info = {
    'quiet': True,
    'extract_flat': True,
    'cookiefile': 'cookies.txt'  # Netscape cookies file
}

# Initialize common download options; more specific options will be set inside the loop
ydl_opts_download = {
    'format': 'worst',  # Download the smallest size available
    'cookiefile': 'cookies.txt'  # Netscape cookies file
}

# Loop through video_ids to download the videos
for video_id in video_ids:
    if counter >= max_videos:
        break

    video_url = f"https://www.youtube.com/watch?v={video_id}"

    try:
        # Decide if video is exciting or not
        exciting = exciting_scores[video_id] > 0.5

        # Set the download path
        download_folder = 'exciting' if exciting else 'non_exciting'

        # Download the video using pytube
        yt = YouTube(video_url)
        video = yt.streams.get_highest_resolution()
        video.download(output_path=download_folder)

        counter += 1
        print(f"Downloaded {counter} videos.")

    except Exception as e:
        print(f"An error occurred for video: {video_id}")
        print(str(e))
        continue
```
